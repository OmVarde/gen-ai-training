```python
import nltk
nltk.download('wordnet')
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "written", "programming", "programs", "programmed"]
```

```python
## QA, Chatbots , text summarization, text generation, sentiment analysis, machine translation, and more.

from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
# POS - Noun:n
# Verb:v
# Adjective:a
# adverb:r


lemmatizer.lemmatize("going", pos="v")
```

```python
for word in words:
    print(word + "---->" + lemmatizer.lemmatize(word, pos="v") )
```

```python
lemmatizer.lemmatize("fairly"), lemmatizer.lemmatize("sportingly")
```
