```python
import pandas as pd
messages = pd.read_csv(r'C:\Users\varde\OneDrive\Desktop\GEN AI\Text Pre Processing\BOW\spam_ham_dataset_5000.csv',
            sep = ',',names = ['message','label'])
```

```python
messages
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## creating bag of words
import sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=84, binary=True)
X = cv.fit_transform(corpus).toarray()
```

```python
# with n - grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=84, binary=True, ngram_range=(2,2))
X = cv.fit_transform(corpus).toarray()
```
