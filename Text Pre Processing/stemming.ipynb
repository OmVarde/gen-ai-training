```python
## stemming --> removes unwanted words that repeats 
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "written", "programming", "programs", "programmed"]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print( word + " --->" + stemming.stem(word) )
```

```python
stemming.stem("congratulations")
```

```python
## here we can see that the word "congratulations" is reduced to "congratul" this is the major disadvantage of stemming .
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|ed$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print( word + " --->" + snowball_stemmer.stem(word) )
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
snowball_stemmer.stem("fairly"), snowball_stemmer.stem("sportingly")
```
